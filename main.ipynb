```python
import numpy as np
import os
import hmm
```

```python
hmm_models = []
for i in range (1, 10):
    hmm_models.append(hmm.HMM(para_path=f"model/model_{i}/hmm_params.npz"))
```

```python
# file = "train_set/train_1/1hao_luo_mfcc.csv"
file = "train_set/train_1/1_mfcc.csv"
mfcc = np.loadtxt(file, delimiter=",")
```

```python
scores = np.zeros(9)
for idx, hmm in enumerate(hmm_models):
    scores[idx] = hmm.Viterbi(mfcc)
print(np.argmax(scores) + 1)
```
