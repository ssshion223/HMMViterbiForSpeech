# train module uses hmmlearn library since there's no need to implement it on fpga

```python
import hmmlearn.hmm
import numpy as np
```

```python
# config
model_id = 9
num_train_set = 6
N = 5
##
model = hmmlearn.hmm.GaussianHMM(
    n_components=N,
    covariance_type="diag",
    n_iter=100
)
```

```python

mfcc_set = []
lengths = []
for i in range(num_train_set):
    mfcc_in = np.loadtxt(f"train_set/spk0{i+1}_mfcc/{model_id}_spk0{i+1}_mfcc.csv", delimiter=",")
    lengths.append(len(mfcc_in))
    mfcc_set.append(mfcc_in)
X = np.vstack(mfcc_set)
model.fit(X, lengths)
# decoding

_B_sigma = np.zeros((N, 12))
for i in range(N):
    for j in range(12):
        _B_sigma[i, j] = model.covars_[i, j, j]
np.savez(
    f"model/model_{model_id}_params.npz",
    pi=model.startprob_,
    A=model.transmat_,
    B_mu=model.means_,
    B_sigma=_B_sigma
)
print(_B_sigma)
```
