# train module uses hmmlearn library since there's no need to implement it on fpga

```python
import hmmlearn.hmm
import numpy as np
```

```python
model = hmmlearn.hmm.GaussianHMM(
    n_components=3,
    covariance_type="diag",
    n_iter=100
)
```

```python
mfcc1 = np.loadtxt("train_set/train_1/1_mfcc.csv", delimiter=",")
mfcc2 = np.loadtxt("train_set/train_1/1hao_luo_mfcc.csv", delimiter=",")
# mfcc3 = np.loadtxt("", delimiter=",")
# mfcc4 = np.loadtxt("", delimiter=",")
# mfcc5 = np.loadtxt("", delimiter=",")

X = np.vstack([mfcc1, mfcc2])
lengths = [len(mfcc1), len(mfcc2)]
model.fit(X, lengths)
# decoding

_B_sigma = np.zeros((3, 12))
for i in range(3):
    for j in range(12):
        _B_sigma[i, j] = model.covars_[i, j, j]
np.savez(
    "model/model_1/hmm_params.npz",
    pi=model.startprob_,
    A=model.transmat_,
    B_mu=model.means_,
    B_sigma=_B_sigma
)
print(_B_sigma)
```
